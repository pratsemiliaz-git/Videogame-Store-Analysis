## 1. Estudio general de los datos

Se importarán las biblitoecas requeridas para poder procesar los dtos, y se 
descargrá el archivo que tiene los datos de años anteriores de las ventas de Ice, para poder dar un vistazo general a su información.



```python
import pandas as pd
import numpy as np 
from scipy import stats as st
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ice_df = pd.read_csv('/Users/RPA/Downloads/games.csv')
print(ice_df.info())
```

```python
ice_df.head()
```

A primera vista se observan un par de cosas que hay que limpiar antes de poder analizar los datos:
- Los nombres de las columnas se pasarán a minúsculas
- La columna 'platform' se convertirá a string en vez de object
- La columna 'year_of_release' se convertira a int64 en vez de float, y se eliminarán las filas con valores nulos en esta columna
- La columna 'genre' se convertirá a string en vez de object
- Se convertirán los valores nulos de las columnas 'critic_score' y 'user_score'
- Se considerará rellenar los valores nulos de 'critic_score' y 'user_score' con un promedio, tomando en cuenta que hay valores "tbd"
- Los valores nulos de 'rating' se convertirán a un nombre genérico como 'N/A', ya que no se les puede asignar un valor si no viene de la ESRB.

## 2. Preparación de los Datos

### Nombres de las Columnas:

```python
col_names = []
for column_name in ice_df.columns:
    column_name = column_name.lower()
    col_names.append(column_name)
ice_df.columns = col_names
print(ice_df.info())
```

Ahora todas las columnas están escritas en minúsculas

### Columna 'name'

Se tratarán los valores nulos y/o vacíos

```python
ice_df['name'].isna().sum()
ice_df['name'] = ice_df['name'].fillna('Unknown')
print(ice_df.info())
```

```python
ice_df.duplicated().sum()
```

La columna name ahora está limpia: no hay valores duplicados y ya no hay valores nulos

### Columna 'platform'

Se tratarán los valores duplicados, y se convertirá a la columna a tipo 'string' en vez de 'object'

```python
print(ice_df.duplicated().sum())
ice_df['platform'] = ice_df['platform'].astype('string')
print(ice_df.info())
```

Sabíamos que 'platform' no tiene valores nulos; los valores duplicados no se van a eliminar por la naturaleza de los datos de esta columna (varios juegos se pueden usar en la misma plataforma).

### Columna 'Year_of_Release'

Se tratarán los valores nulos y/o duplicados, y se cambiará el formato de los valores de float64 a int64

```python
print(ice_df['year_of_release'].unique())
ice_df = ice_df.dropna(subset=['year_of_release'])
print(ice_df['year_of_release'].isna().sum())
ice_df['year_of_release'] = ice_df['year_of_release'].astype('int')
print(ice_df.info())
print(ice_df['year_of_release'].unique())
```

Se eliminaron todas las filas del DF donde 'year_of_release' tuviera valores nulos, para mejor manejo de los datos, así como facilitar los cálculos futuros. No se eliminó ningún valor duplicado, porque varios juegos pudieron haber salido al aire en el mismo año.

### Columna 'Genre'

Se rellenarán los valores nulos. El formato de los valores se dejará como 'object', porque hay géneros que llevan tildes, guiones u otros caractéres. 

```python
ice_df['genre'] = ice_df['genre'].fillna('Unknown')
```

La columna 'genre' ahora tiene todos sus valores como no-nulos. 

### Columnas 'Sales'

No se procesarán las columnas us_sales, eu_sales, jp_sales ni other_sales. Ninguna tiene valores nulos, y su formato se dejará tal y como está en float. No se hará nada en cuanto a potenciales valores duplicados. 

### Columnas 'Score'

Se procesarán las dos columnas de 'score'. Se abordarán los valores nulos.

Se cambiará el formato de datos de la columna 'user_score' de object a float.

Al inicio, se consideró dejar los valores nulos de estas columnas como eso, valores nulos. Pero una vez que se alcanzó la sección de estadística, me di cuenta que es mejor opción eliminar las filas que tengan valores nulos en estas columnas. Se hará lo mismo con las que tienen "tbd". Esto se hará más adelante, para no afectar los otros cálculos del proyecto.

```python
ice_df['user_score'] = ice_df['user_score'].replace('tbd','NaN')
ice_df['user_score'] = ice_df['user_score'].astype('float')
print(ice_df.head())
print(ice_df.info())
```

Critic_score se dejará tal cual está

### Columna 'Rating'

Por último, en esta columna también se abordarán los valores nulos.

```python
print(ice_df.info())
print(ice_df['rating'].unique())
print(ice_df.head(), ice_df.head(-5))
```

No se hizo nada en cuanto a los valores nulos de esta columna.

Se puede observar que la mayoría de las filas que tienen valores nulos en 'rating' también tienen valores nulos en 'critic_score' y 'user_score'.

Se puede deducir que loas celdas con valores nulos o 'tbd' en las columnas de score están así porque hay países en los que algunos videojuegos no se vendieron, o porque el ESRB todavía no les brinda una calificación, pero esto es poco probable porque algunos de los juegos que tienen 'tbd' en 'rating' salieron desde hace varios años, y no tendría sentido que la ESRB todavía no los califique para el público. Dependiendo de pasos más adelante, se contemplará eliminar las filas que tengan valores nulos en estas columnas.

### Ventas totales

Se hará el cálculo de ventas totales en NA, EU y JP y se agregará como una nueva columna del DF. 

```python
ice_df['total_sales'] = ice_df[['na_sales','eu_sales','jp_sales']].sum(axis=1)
print(ice_df.info())
```

### 3.2 - Observa cómo varían las ventas de una plataforma a otra

Elige las plataformas con las mayores ventas totales y construye una distribución basada en los datos de cada año. Busca las plataformas que solían ser populares pero que ahora no tienen ventas. ¿Cuánto tardan generalmente las nuevas plataformas en aparecer y las antiguas en desaparecer?

Observaremos las ventas totales de cada plataforma

```python
sales_by_platform = ice_df.groupby('platform')['total_sales'].sum().sort_values(ascending=False).reset_index()
sales_by_platform.plot(kind='bar', title='Ventas Históricas por Plataforma', ylabel='Ventas (MDD)', x='platform', xlabel='Platform', ylim=[0,1101], grid=True, figsize=[8,6])
print(ice_df['platform'].nunique())
print(sales_by_platform)
```

Con los cálculos anteriores, podemos ver que hay 6 plataformas que tienen ventas mucho mayores que las demás, con 'PS2' siendo la plataforma con mayores ventas, seguido por 'X360'. Para los siguientes análisis, consideraremos las plataformas que han vendido al menos más de 200 MDD.

Elegiremos  las plataformas con mayores ventas totales y las graficaremos

```python
top_platforms = sales_by_platform.head(10)['platform']
biggest_platforms = ice_df[ice_df['platform'].isin(top_platforms)]
pivot_sales = biggest_platforms.pivot_table(columns='platform', index='year_of_release', values='total_sales', aggfunc = 'sum')
pivot_sales.plot(kind = 'line', title='Plataformas con Mayores Ventas (Ventas por Año)', ylabel='Ventas(MDD)', xlabel='Año', rot=90, figsize=[8,6], marker = 'o', grid='True', ylim=[0,201])
plt.xticks(ticks=pivot_sales.index)
plt.legend(title='Plataformas')
plt.show()
```

Conclusiones que se pueden hacer partiendo de esta gráfica: 
- La plataforma que más ha vendido en un solo año es Wii, con cerca de 190 MDD en 2009; seguida por PS2 con cerca de 180 MDD en 2002.
- Wii también tuvo una de las caídas más drásticas de todas las plataformas: pasó de vender casi 190 MDD en 2009 a casi 50 MDD en 2011. 
- La mayoría de las plataformas tiende a obtener un punto máximo de ventas y después disminuir. Es decir, no es común que la plataforma mantenga sus ventas iguales durante varios años.
- La plataforma más irregular en cuanto a subida y bajada de ventas fue GameBoy entre los años 1988 y 2001, haciendo prácticamente una subida y una bajada por año. Aún así, es la plataforma que más tiempo ha pasado en el mercado.
- Viendo esta gráfica, podríamos decir que los datos de los años anteriores a 1994 son irrelevantes para nuestro análisis, ya que hay muy pocos datos significativos en esos años. Para análisis posterior, se eliminarán estos años.
- La cantidad de años promedio que una plataforma tiene sus ventas más altas es de 3-4 años antes de bajar sus ventas.
- La bajada de ventas de una plataforma casi siempre coincide con la publicación de una nueva plataforma. Por ejemplo, las ventas de PS comienzan a bajar en el año 2000, que es el mismo año en el que se lanzó el PS2, que también tuvo su bajada en el año 2006, que es cuando salieron Wii y PS3.

### 3.4 Mira cómo las reseñas de usuarios y profesionales afectan las ventas de una plataforma popular (tu elección). Crea un gráfico de dispersión y calcula la correlación entre las reseñas y las ventas. Saca conclusiones.

Se puede decir que si hay cierta influencia entre la calificación y las ventas, pero no parece ser un punto de partida para la decisión de compra de los usuarios. La variable que muestra un poco más de correlacion con las ventas es la calificación de los profesionales. 

Intentemos con otra plataforma: 

```python
wii_sales= ice_df[ice_df['platform']=='Wii']
wii_user_corr = wii_sales['total_sales'].corr(wii_sales['user_score'])
print('Correlación entre reseñas de usuarios y ventas de Wii: ', wii_user_corr)
wii_critic_corr = wii_sales['total_sales'].corr(wii_sales['critic_score'])
print('Correlación entre reseñas de críticos y ventas de Wii: ', wii_critic_corr)
wii_usuarios = wii_sales[['user_score', 'total_sales']]
wii_criticos = wii_sales[['critic_score', 'total_sales']]
pd.plotting.scatter_matrix(wii_usuarios)
pd.plotting.scatter_matrix(wii_criticos)
```

### 3.5 Compara las ventas de los mismos juegos en diferentes plataformas

## 5. Formulación de Hipótesis
Probemos un par de hipótesis sobre nuestros datos analizados

####  H0: Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son las mismas
#### H1: Las calificaciones promedio de los usuarios de las plataformas Xbox One y PC son diferentes

Primero se eliminarán los datos NaN y tbd de la columna 'user_score':

```python
ice_df = ice_df.dropna(subset=['user_score'])
print(ice_df.head())
print(ice_df.info())
```

Ahora se comprobará si la H0 se rechaza o no: 

```python
xbox_platform = ice_df[ice_df['platform']=='XOne']
pc_platform = ice_df[ice_df['platform']=='PC']
sample_xbox = xbox_platform['user_score']
sample_pc = pc_platform['user_score']
alpha = 0.05
result= st.ttest_ind(sample_xbox, sample_pc)
print('Valor p:' , result.pvalue)
if result.pvalue < alpha:
    print('Rechazamos la hipótesis nula')
else:
    print('No podemos rechazar la hipótesis nula')
```

La Hipótesis nula es rechazada, lo cual significa que las calificaciones promedio de los usuarios sobre las plataformas Xbox One y PC no son iguales. 

### H0: Las calificaciones promedio de los usuarios para los géneros de Acción y Deportes son diferentes

### H1: Las calificaciones promedio de los usuarios para los géneros de Acción y Deportes no son diferentes

```python
action = ice_df[ice_df['genre']=='Action']
sport = ice_df[ice_df['genre']=='Sports']
action_score = action['user_score']
sport_score= sport['user_score']
alpha = 0.05
result = st.ttest_ind(action_score, sport_score)
print('valor p:' , result.pvalue)
if result.pvalue < alpha:
    print('Rechazamos la hipótesis nula')
else: 
    print('No podemos rechazar la hipótesis nula')
```

No pudimos rechazar la hipótesis nula. Lo cual significa que (aunque no con 100% de certeza), las calificaciones promedio de los usuarios para los géneros de acción y deportes son diferentes.
